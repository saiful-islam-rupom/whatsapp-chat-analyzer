```python
import re
import pandas as pd
import numpy as np
```

```python
f = open("D:\whatsapp-chat-analyzer\data\WhatsApp Chat with IELTS E2 Batch.txt",'r',encoding='utf-8')
```

```python
data= f.read()
```

```python
print(data)
```

```python
print(type(data))
```

```python
pattern = r'\d{1,2}/\d{1,2}/\d{2,4}, \d{1,2}:\d{2}(?:\u202F|\u00A0| )?(?:am|pm) - '
```

```python
messages = re.split(pattern,data)[1:]
```

```python
messages
```

```python

def clean_m():
    cleaned = [d.replace("\u202f", " ").replace("\u00A0", " ").strip() for d in messages]
    return cleaned

clean_messages = clean_m()


clean_messages
```

```python
dates = re.findall(pattern,data)
dates
```

```python
def clean_d():
    cleaned = [d.replace("\u202f", " ").replace("\u00A0", " ").strip() for d in dates]
    return cleaned

clean_dates = clean_d()


clean_dates
```

```python
df = pd.DataFrame({'user_message':clean_messages, 'message_date':clean_dates})

df
```

```python

df['message_date'] = df['message_date'].str.replace(r'\s*-\s*$', '', regex=True)

df['message_date'] = pd.to_datetime(
    df['message_date'].str.replace(r'\s*-\s*$', '', regex=True),
    format='%m/%d/%y, %I:%M %p'
)

df.rename(columns={'message_date': 'date'}, inplace= True)

df['date'] = df['date'].dt.strftime(r'%m/%d/%y, %I:%M %p')

df.head(20)
```

```python
df.shape
```

```python
# separate user and messages
users = []
messages = []

for message in df['user_message']:
    entry = re.split(r'^(.*?):\s', message, maxsplit=1)
    if len(entry) >= 3:  # has a user
        users.append(entry[1])     # the user (name or phone)
        messages.append(entry[2])  # the actual message
    else:  # system notification (no user)
        users.append('group_notification')
        messages.append(entry[0])

df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace= True)

df.head(20)
```

```python
df['date'] = pd.to_datetime(
    df['date'],
    format='%m/%d/%y, %I:%M %p'
)
```
